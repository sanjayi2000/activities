```python
 #Create a bar graph for countrywise distribution with reduced font size for country names
plt.figure(figsize=(12, 6))
bars = plt.bar(country_distribution.index, country_distribution.values)
plt.title('Countrywise Distribution of Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xticks(fontsize=6)  # Reduce font size of country names
plt.tight_layout()  # Adjust spacing to prevent overlap
plt.show()
```

```python
merged_df = md.merge(df, on='country_code')

# Ensure data cleaning and consistency
merged_df = merged_df.dropna(subset=['country_code', 'medal_type'])  # Handle missing values
merged_df = merged_df.drop_duplicates()  # Remove duplicates

# Define a function to create subplots for each country
def plot_medals_by_country(data):
  countries = data['country'].unique()
  fig, axes = plt.subplots(nrows=len(countries), figsize=(10, len(countries)*3))

  for i, country in enumerate(countries):
    country_data = data[data['country'] == country]
    medal_counts = country_data['medal_type'].value_counts()
    medal_counts.plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'{country} Medals')
    axes[i].set_xlabel('Medal Type')
    axes[i].set_ylabel('Count')

  plt.tight_layout()
  plt.show()

# Call the function to create subplots
plot_medals_by_country(merged_df.copy())
```
